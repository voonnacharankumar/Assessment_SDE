```python
# Reading document using pandas
import pandas as pd
from datetime import timedelta

df = pd.read_excel('Assignment_Timecard.xlsx')

# Dropping unnecessary columns

df.drop(['Unnamed: 9', 'File Number', 'Unnamed: 10','Position Status','Pay Cycle End Date','Pay Cycle Start Date'], axis=1, inplace=True)

# Dropping rows which have NaN values

df.dropna(subset=['Time', 'Time Out'],inplace=True)

#Creating Date column

df['Date'] = df['Time'].dt.date

#Sorting data according to Employee Name and Time 

df.sort_values(by=['Employee Name', 'Time'], inplace=True)

#Calculating shift duration and converting to hours

df['Shift Duration'] = (df['Time Out'] - df['Time']).dt.total_seconds() / 3600


print("-----who has worked for 7 consecutive days-----")

consecutive_workdays = 7
consecutive_7_workdays=set()
#Checking according to the Employer Name using groupby
for name, group in df.groupby('Employee Name'):
    
    #storing dates of the employees
    unique_dates = group['Date'].unique()

    # Check for consecutive workdays
    for i in range(len(unique_dates) - consecutive_workdays + 1):
        consecutive_days = unique_dates[i:i + consecutive_workdays]
        if (consecutive_days[-1] - consecutive_days[0]).days == consecutive_workdays - 1:
            position = group.loc[group['Date'].isin(consecutive_days), 'Position ID'].iloc[0]
            consecutive_7_workdays.add(f"Employee Name: {name}, Position: {position}")
            break
for x in consecutive_7_workdays:
    print(x)

print()
print()
print("--who have less than 10 hours of time between shifts but greater than 1 hour--")

time_shift_btw_1_to_10=set()

for name, group in df.groupby('Employee Name'):
    sorted_group = group.sort_values(by='Time')
    for i in range(len(sorted_group) - 1):
        present_shift_end_time = sorted_group.iloc[i]['Time Out']
        next_shift_start_time = sorted_group.iloc[i + 1]['Time']
        time_diff = next_shift_start_time - present_shift_end_time
        if timedelta(hours=1) < time_diff < timedelta(hours=10):
             time_shift_btw_1_to_10.add(f"Employee Name: {name}, Position: {sorted_group.iloc[i]['Position ID']}")
              break

for x in time_shift_btw_1_to_10:
    print(x)

    
print()
print()
print(" Who has worked for more than 14 hours in a single shift")

worked_morethan_14hrs=set()

# Identify employees who have worked for more than 14 hours in a single shift
for name, group in df.groupby('Employee Name'):
    for index, row in group.iterrows():
        if row['Shift Duration'] > 14:
            worked_morethan_14hrs.add(f"Employee Name: {row['Employee Name']}, Position: {row['Position ID']}")
            break

for x in worked_morethan_14hrs:
    print(x)
```
